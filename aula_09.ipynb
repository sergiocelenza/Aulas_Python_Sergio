# Aula 09: Funções

Até hoje, nós escrevemos o código de cima para baixo. Mas e se precisarmos repetir um mesmo bloco de código em várias partes do programa (por exemplo, limpar a tela e mostrar um menu)?
As **Funções** nos permitem dar um *nome* para um bloco de código, para que possamos "chamá-lo" quando quisermos, ajudando a organizar nosso programa.

## 1. Criando uma Função

Para "definir" uma função, usamos a palavra `def`, seguida do nome da função e parênteses `()`.

```python
def dar_boas_vindas():
    print("----------------------")
    print(" Bem vindo ao Sistema ")
    print("----------------------")

# A função não faz nada até ser explicitamente "chamada"
print("Iniciando...")
dar_boas_vindas()
print("Fazendo algo no sistema...")
dar_boas_vindas()
```

## 2. Passando Parâmetros (Argumentos)

Funções podem receber dados de fora para trabalhar com eles. Chamamos isso de "parâmetros".

```python
def saudar_pessoa(nome_da_pessoa):
    print(f"Olá, {nome_da_pessoa}! Seja muito bem-vindo.")


saudar_pessoa("Maria")
saudar_pessoa("Pedro")
```

Podemos ter múltiplos parâmetros, separados por vírgula.

```python
def somar(a, b):
    resultado = a + b
    print(f"A soma de {a} com {b} é {resultado}")

somar(10, 5)
somar(100, 200)
```

## 3. Retornando Valores (`return`)

A função `somar` acima imprimiu o valor na tela. Mas e se quiséssemos pegar esse resultado para usar em outra conta? Funções podem **devolver** um resultado para quem as chamou usando a palavra `return`.

```python
def multiplicar(a, b):
    resultado = a * b
    return resultado # "Devolve" o valor, mas não imprime!

# Guardamos o retorno em uma variável
dobro_de_10 = multiplicar(10, 2)
print(f"O retorno da função foi {dobro_de_10}")

# Assim podemos usar o resultado em outra parte do código
print(f"O quádruplo de 10 é {dobro_de_10 * 2}")
```

## 4. Retornando Múltiplos Valores (Tuplas)

Uma característica incrível e muito usada no Python é a capacidade de uma função retornar **mais de um valor ao mesmo tempo**. Diferente de linguagens como C ou Java (onde você normalmente tem que fazer truques ou criar classes para devolver dois valores), no Python basta separá-los por vírgula no `return`.

```python
def calcular_estatisticas(numeros):
    # Calcula três coisas de uma vez
    total = sum(numeros)
    maior = max(numeros)
    menor = min(numeros)

    # Retornamos os três valores separados por vírgula.
    # Isso magicamente os agrupa no que o Python chama de 'Tupla' (parecida com uma lista, mas imutável)
    return total, maior, menor

minhas_notas = [7, 8, 10, 5, 9]

# Na hora de chamar a função, também usamos variáveis separadas por vírgula para receber as 3 respostas!
# Chamamos isso de 'Desempacotamento' (Unpacking)
soma_total, nota_maxima, nota_minima = calcular_estatisticas(minhas_notas)

print(f"Soma: {soma_total}")
print(f"Máxima: {nota_maxima}")
print(f"Mínima: {nota_minima}")
```

Se você guardar a resposta numa variável só (sem desempacotar), o Python te devolverá os itens entre parênteses `()`. Isso é o que define uma **Tupla**.

```python
tudo_junto = calcular_estatisticas(minhas_notas)
print("Retorno direto (Tupla):", tudo_junto)

# Tuplas funcionam igual a listas para acessar pelo índice, mas você não pode alterá-las depois de criadas:
print("Acessando o primeiro item da tupla (soma):", tudo_junto[0])
```

# Lista de Exercícios Extra: Funções, Listas e Loops

Pratique os conceitos combinados abaixo:

### 1. Soma de Lista
Crie uma função `somar_elementos(lista)` que recebe uma lista de números e retorna a soma de todos eles (não use a função `sum()`).

```python
def somar_elementos(lista):
    soma = 0
    for num in lista:
        soma += num
    return soma

# Teste
print(somar_elementos([1, 2, 3, 4, 5]))
```

### 2. Média de Notas
Crie uma função `calcular_media(notas)` que recebe uma lista de notas e retorna a média aritmética.

```python
def calcular_media(notas):
    if not notas: return 0
    return sum(notas) / len(notas)

# Teste
print(calcular_media([10, 8, 9, 7]))
```

### 3. Apenas Pares
Crie uma função `filtrar_pares(numeros)` que recebe uma lista e retorna uma **nova lista** contendo apenas os números pares.

```python
def filtrar_pares(numeros):
    pares = []
    for n in numeros:
        if n % 2 == 0:
            pares.append(n)
    return pares

# Teste
print(filtrar_pares([1, 2, 3, 4, 5, 6]))
```

### 4. Maior Valor
Crie uma função `achar_maximo(lista)` que percorre a lista e retorna o maior valor (não use `max()`).

```python
def achar_maximo(lista):
    if not lista: return None
    maior = lista[0]
    for item in lista:
        if item > maior:
            maior = item
    return maior

# Teste
print(achar_maximo([15, 8, 22, 11]))
```

### 5. Contador de Palavras Curtas
Crie uma função `contar_curtas(palavras)` que recebe uma lista de strings e retorna quantas têm menos de 5 letras.

```python
def contar_curtas(palavras):
    contador = 0
    for p in palavras:
        if len(p) < 5:
            contador += 1
    return contador

# Teste
print(contar_curtas(["sol", "python", "casa", "algoritmo"]))
```

### 6. Tabuada em Lista
Crie uma função `gerar_tabuada(n)` que retorna uma lista com a tabuada de `n` (de 1 a 10).

```python
def gerar_tabuada(n):
    resultados = []
    for i in range(1, 11):
        resultados.append(n * i)
    return resultados

# Teste
print(gerar_tabuada(7))
```

### 7. Inversor de Lista
Crie uma função `inverter_lista(lista)` que retorna a lista na ordem inversa (não use `.reverse()`).

```python
def inverter_lista(lista):
    invertida = []
    for i in range(len(lista) - 1, -1, -1):
        invertida.append(lista[i])
    return invertida

# Teste
print(inverter_lista([1, 2, 3]))
```

### 8. Fatorial com Loop
Crie uma função `fatorial(n)` que calcula o fatorial de um número usando um laço `for` ou `while`.

```python
def fatorial(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

# Teste
print(fatorial(5))
```

### 9. Verificador de Presença
Crie uma função `esta_na_lista(lista, elemento)` que retorna `True` se o elemento estiver na lista e `False` caso contrário.

```python
def esta_na_lista(lista, elemento):
    for item in lista:
        if item == elemento:
            return True
    return False

# Teste
print(esta_na_lista(["Maçã", "Banana", "Uva"], "Banana"))
```

### 10. Multiplicar Lista por Escalar
Crie uma função `multiplicar_lista(lista, fator)` que retorna uma nova lista onde cada item da original é multiplicado pelo fator.

```python
def multiplicar_lista(lista, fator):
    nova_lista = []
    for item in lista:
        nova_lista.append(item * fator)
    return nova_lista

# Teste
print(multiplicar_lista([1, 2, 3], 10))
```
